# EDA --> Explory Data Analysis

```python
df['State'].nunique()
```

```python
df.sort_values(by='Active_Ratio',ascending=False)
```

```python
df.groupby(['State'])['Active_Ratio'].max().sort_values(ascending=False).reset_index()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active_Ratio'],color='red',marker='o')
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State and UT', fontsize=15)
plt.ylabel('Active Ratio', fontsize=15)
plt.title("Line Plot for Active Ratio",fontsize=15,loc='center')
plt.show()
```

```python
total_case=df.sort_values(by='Total_Cases',ascending=False)
```

```python
total_case
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='State',y='Total_Cases',data=total_case)
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State and UT', fontsize=15)
plt.ylabel('Total_Cases', fontsize=15)
plt.title("Bar Plot for Total_Cases",fontsize=15,loc='center')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='State',y='Total_Cases',data=total_case.head())
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State and UT', fontsize=15)
plt.ylabel('Total_Cases', fontsize=15)
plt.title("Bar Plot for Total_Cases For Top-5 States",fontsize=15,loc='center')
plt.show()
```

## **Maharashtra has the highest number of total cases in India, Kerala and Karnataka rank second and third although there is not much difference between their total counts. *

```python
plt.figure(figsize=(10,4))
sns.barplot(x='State',y='Total_Cases',data=total_case.tail())
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State and UT', fontsize=15)
plt.ylabel('Total_Cases', fontsize=15)
plt.title("Bar Plot for Total_Cases for least-5 States",fontsize=15,loc='center')
plt.show()
```

## *Andaman and Nicobar has the lowest number of total cases in India, Lakshadweep and DANHADAD rank second and third although there is not much difference between their total counts. *

```python
plt.figure(figsize=(10,4))
sns.barplot(x='State',y='Total_Cases',data=total_case.head(10),palette='Set2')
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State and UT', fontsize=15)
plt.ylabel('Total_Cases', fontsize=15)
plt.title("Bar Plot for Total_Cases for top-10 States",fontsize=15,loc='center')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.barplot(x='State',y='Total_Cases',data=total_case.tail(10),palette='Set1')
plt.xticks(rotation=90,fontsize=11)
plt.yticks(fontsize=13.5)
plt.xlabel('State and UT', fontsize=15)
plt.ylabel('Total_Cases', fontsize=15)
plt.title("Bar Plot for Total_Cases for least-10 States",fontsize=15,loc='center')
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Line Plot for Active Ratio and Death Ratio",fontsize=15)

plt.plot(df['State'],df['Active_Ratio'],color='red',marker='o',label='Active cases')
plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)
plt.xlabel('State', fontsize=15)
# plt.legend()

plt.ylabel('Active Ratio', fontsize=15)

plt.plot(df['State'],df['Death_Ratio'],color='blue',marker='*',label='death')
plt.legend()
plt.show()
```

```python
df['Recovered_case']=df['Total_Cases']-(df['Active']+df['Deaths'])
```

```python
df
```

```python
df['Recovered_ratio']=df['Recovered_case']/df['Total_Cases']*100
```

```python
df
```

```python
df.head(2)
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active_Ratio'],color='blue',marker='o',label='Active_Ratio')
plt.plot(df['State'],df['Death_Ratio'],color='k',marker='*',label='Death_Ratio')
plt.plot(df['State'],df['Recovered_ratio'],color='green',marker='*',label='Recovered_ratio')

plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)
plt.xlabel('State', fontsize=15)


plt.ylabel('Covid Ratio', fontsize=15)
plt.title("Line Plot for Deth, Active and Recoverd Ratio",fontsize=15)


plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['State'],df['Active'],color='blue',marker='o',label='Active_Cases')
plt.plot(df['State'],df['Deaths'],color='red',marker='*',linestyle='--',label='Death_cases')
plt.plot(df['State'],df['Recovered_case'],color='green',marker='*',label='Recovered_case')

plt.xticks(rotation=90,fontsize=13.5)
plt.yticks(fontsize=13.5)
plt.xlabel('State', fontsize=15)


plt.ylabel('Covid Cases', fontsize=15)
plt.title("Line Plot for Deth, Active and Recoverd Cases",fontsize=15)


plt.legend()
plt.show()
```
